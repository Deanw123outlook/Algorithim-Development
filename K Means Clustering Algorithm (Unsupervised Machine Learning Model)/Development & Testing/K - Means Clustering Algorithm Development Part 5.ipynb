```python
#import required libraries
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from math import sqrt
```

```python
print('Please input the number of elements required :')
n1 = input()
#convert string to integer rationale range function & randinint requires interger value
n1 = int(n1)
print('Please input the max range :')
n2 = input()
n2 = int(n2)

#create two random lists to hold X & Y co - ordinates of elements
#X co-ordinates
X = []
#Y co - ordinates
Y = []

#intialize for loop with the range function (start/end)
for i in range(0,n1):
    #Generates a random number between a given positive range
    x = random.randint(1,n2)
    #appends to empty list(adds)
    X.append(x)
for i in range(0,n1):
    y = random.randint(1,n2)
    Y.append(y)

#Initiate Centroid 1
print('Please enter centroid 1 co - ordinates')
print('Enter x co - ordinate')
cx1 = input()
cx1 = int(cx1)
print('Enter y co - ordinate')
cy1 = input()
cy1 = int(cy1)
print('Centroid 1 x co - ordinates:',cx1)
print('Centroid 1 y co - ordinates:',cy1)

#Initiate Centroid 2
print('Please enter centroid 2 co - ordinates')
print('Enter x co - ordinate')
cx2 = input()
cx2 = int(cx2)
print('Enter y co - ordinate')
cy2 = input()
cy2 = int(cy2)
print('Centroid 2 x co - ordinates:',cx2)
print('Centroid 2 y co - ordinates:',cy2)

#SCATTERPLOT 
#x & y co-ordniates
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.scatter(X, Y)
# add centroid 1 & centroid 2
plt.scatter(x=[cx1], y=[cy1], color='b')
plt.scatter(x=[cx2], y=[cy2], color='g')
plt.title("X & Y Co-Ordinates | Centroid 1 & 2")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Initiate Centroid Values (For Re-Assignment Purposes)
#These values will change as the program runs through iterations and reaches a termination threshold
new_centroid1x = 0
new_centroid1y = 0

#Print for Interpretation
print('New Centorid Value', new_centroid1x, new_centroid1y)
print(cx1,cy1)
print(cx2,cy2)
```

```python
#INITIATE ALGORITHIM WHILE LOOP
#The program will continue to re-run until there is no change within centroid 
#co-ordinates (reaching the termination criterion)
while cx1 != new_centroid1x:
    while cy1 != new_centroid1y:
        
        #initiliaze empty lists
        list1 = []
        list2 = []
        list3 = []
        list4 = []
        list5 = []
        list6 = []
        
        #eulcidean distance calculations
        for xi in X:
            #while loop : terminate the for loop when length of list1 exceeds the length of the list2
            while len(X) <= len(X):
                #appending(adding calculations to list)
                list1.append((cx1 - xi) ** 2)
                #appending(adding calculations to list)
                break

        for yi in Y:
            while len(Y) <= len(Y):
                list2.append((cy1 - yi) ** 2)
                break
        
        #for loop to loop over list
        for xii in X:
            while len(X) <= len(X):
                list3.append((cx2 - xii) ** 2)
                break
        
        for yii in Y:
            while len(Y) <= len(Y):
                list4.append((cy2 - yii) ** 2)
                break
        
        #for loop zipping together two lists   
        for x,y in zip (list1,list2):
            while len(list1) <= len(list2):
                #square root calculations
                phase1 = sqrt(x + y)
                list5.append(phase1)
                break
                
        for a,b in zip (list3,list4):
            while len(list3) <= len(list4):
                phase2 = sqrt(a + b)
                list6.append(phase2)
                break
                
        #Transfering the above lists into a Dictionary
        Dictionary = {'Xco-ordinates':X, 'Yco-ordinates':Y, 'EuclideanDistanceCalculationsCentroid1':list5, 
                  'EuclideanDistanceCalculationsCentroid2':list6}
        #Create a DataFrame from our above Dictionary
        df = pd.DataFrame(Dictionary)
        #Create copy of original dataframe
        original_df = df.copy(deep=True)
        
        #Labelling the clusters for interpretability
        Clusters = []
        #Assign Labels for Centroid/Clusters
        label1 = 'Centroid1'
        label2 = 'Centroid2'
        #initiate for loop to assign labels
        #NOTE: These labels will represent which Cluster X & Y co-ordinates belong too !
        for i in df['EuclideanDistanceCalculationsCentroid1']:
            for f in df['EuclideanDistanceCalculationsCentroid2']:
                if i > f:
                    Clusters.append(label1)
                if i <= f:
                    Clusters.append(label2)
                break
                
        #Append to Dataframe
        df['ClusterLabels'] = Clusters
        print(df)
        
        #FINDING NEW CENTROIDS (ITERATIONS)
        
        #isolate based on centroids
        #centroid1
        Centroid1 = df.loc[df['ClusterLabels'] == 'Centroid1']
        
        #count
        CentroidCount = Centroid1['ClusterLabels'].count()
        
        #Adding all elements of Xco-ordinates
        sum_allx = 0
        for x in Centroid1['Xco-ordinates']:
            sum_allx += x
            new_centroid1x = round(sum_allx / CentroidCount,5)
        print('NewCentroid1 Xco-ordinate :', new_centroid1x)
        
        #Adding all elements of Yco-ordinates
        sum_ally = 0
        for y in Centroid1['Yco-ordinates']:
            sum_ally += y
            new_centroid1y = round(sum_ally / CentroidCount,5)
        print('NewCentroid1 Yco-ordinates :', new_centroid1y)
        
        #isolate based on centroids
        #centroid2
        Centroid2 = df.loc[df['ClusterLabels'] == 'Centroid2']
        #count
        CentroidCount = Centroid2['ClusterLabels'].count()
        
        #Adding all elements of Xco-ordinates
        sum_allx = 0
        for x in Centroid2['Xco-ordinates']:
            sum_allx += x
            new_centroid2x = round(sum_allx / CentroidCount,5)
        print('NewCentroid2 Xco-ordinate :', new_centroid2x)
        
        #Adding all elements of Yco-ordinates
        sum_ally = 0
        for y in Centroid2['Yco-ordinates']:
            sum_ally += y
            new_centroid2y = round(sum_ally / CentroidCount,5)
        print('NewCentroid2 Yco-ordinates :', new_centroid2y)
        
        #BREAK FROM INITIAL WHILE LOOP
        break

df
```
