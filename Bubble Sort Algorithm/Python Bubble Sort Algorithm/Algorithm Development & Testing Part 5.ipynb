ERROR HANDLING

Error Handling will be a key feature in our code. This will enable the alogirthm to idnetify & handle any errors. We know from previous tests that an error will be generated when there the index is out of range. The aim of the try/except block, is once the error is generated we will reset the p1/p2 (indexer) code, in combination with the while loop will enable us to loop back to start of the list and begin the process again.

NOTE
- The try block lets you test a block of code for errors.
- The except block lets you handle the error.

```python
#Bubble-Sort-Algorithm
while c < 25: # execute while loop if condition is true ie. c less than 25
    c += 1
    print("Original Postion Indexer",p1)
    print("Original Postion Indexer",p2)
    print("")
    #for loop program !
    for x in List_D:
        p1 += 1 # addition assignment operator
        p2 +=1 # addition assignment operator
        print("Updated Postion Index 1:",p1)
        print("Updated Position Index 2:",p2)
        for i in range(p1,p2,1): # range(start,stop,step)
            if p2 <= len(List_D):
                try: # try block lets you test a block of code for errors.
                    print("Loop-Range-Experimentation")
                    print(List_D[p1])
                    print(List_D[p2])
                    if List_D[p1] > List_D[p2]:
                        List_D[p1],List_D[p2] = List_D[p2],List_D[p1]
                        print(List_D)
                        print("")
                except: # except block enable error handling 
                    #This code (try/except) will enable us to reset the index and loop back to start of the list for this alogorithm
                    print("--Index-Out-Of-Range--")
                    print("--Error Handling Required--")
                    p1 = -1 # reseting the index
                    p2 = 0 # reseting the index
    if c == 25: # of statement for output analysis
        print("Bubble-Sort-Algorithm-Status: COMPLETE")
        print("Please see the sorted list")
        print(List_D)
```

We have succesfuly completed the Bubble Sort Algorithm using the above code. Next, we will improve the code & algorithm output readability!
